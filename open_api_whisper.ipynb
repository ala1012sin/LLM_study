```python
def summarize_txt():
    client = OpenAI(api_key = api_key)
    
    audio_path = 'lsy_audio_2023_58s.mp3'

    with open(audio_path, 'rb') as audiofile:
        transcription_eng = client.audio.translations.create(
            model ='whisper-1',
            file = audiofile
        )
    
    prompt = f"""
    다음은 회의의 본문입니다. 아래 내용을 바탕으로 핵심 내용을 회의 요약 형식으로 정리하세요.
    - 회의 목적
    - 구현 방법
    - 주요 결과
    - 결론 및 시사점
    
    --- 본문 ---
    {transcription_eng}"""
    
    response = client.chat.completions.create(
        model="gpt-4o",              
        temperature=0.9,            
        messages = [
            {"role": "system", "content": "너는 회의록 요약 전문가야."},
            {"role": "user", "content": prompt},
        ],
    )
    return response.choices[0].message.content

print(summarize_txt())
```
